````python
def change_string_to_int(column):
    variables=pd.unique(df_copy[column])
    for item in range(variables.size):
        df_copy[column]=[item if each==variables[item] else each for each in df_copy[column]]
    return df_copy[column]
````

````python
object_columns = df.select_dtypes(include=['object'])

for column in object_columns.columns:
    unique_elements = object_columns[column].unique()
    print(f"Unique elements in '{column}':")
    print(unique_elements)
    print("\n")
````

````python
for column in object_columns.columns:
    label_encoder = LabelEncoder()

    numeric = label_encoder.fit_transform(df_copy[column].unique())
    df_copy[column] = label_encoder.transform(df_copy[column])
    # Create a new DataFrame with the original and encoded values
    encoded_df = pd.DataFrame({column: df[column].unique(), column+' Associated Number': numeric})
    print(encoded_df)
````
